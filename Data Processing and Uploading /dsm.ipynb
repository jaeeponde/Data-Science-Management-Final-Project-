```python
import pandas as pd
import json

# Load the list of 100,000 user IDs
users_df = pd.read_csv("/Users/jaeeponde/dsm_project/data/user_core_100k.csv")
user_set = set(users_df['user_id'])

# Prepare for friend edges
friend_edges = set()

# Path to Yelp user JSON
input_path = "/Users/jaeeponde/dsm_project/Yelp JSON/yelp_academic_dataset_user.json"

# Process each line (each user is one JSON object per line)
with open(input_path, 'r') as f:
    for line in f:
        user = json.loads(line)
        uid = user['user_id']
        
        # Only consider this user if they're in the selected 100k
        if uid in user_set:
            # Filter friends who are also in the selected users
            valid_friends = [fid for fid in user['friends'].split(', ') if fid in user_set]
            for fid in valid_friends:
                # Add a sorted tuple to prevent duplicate/undirected entries
                edge = tuple(sorted((uid, fid)))
                friend_edges.add(edge)

# Convert to DataFrame
friends_df = pd.DataFrame(list(friend_edges), columns=['user_id_1', 'user_id_2'])

# Save to CSV
friends_df.to_csv("/Users/jaeeponde/dsm_project/user_core_100k_friend_edges.csv", index=False)

print(f"✅ Saved {len(friends_df)} unique mutual friend edges.")
```

```python
import json
import csv

# File paths
user_path = '/Users/jaeeponde/dsm_project/Yelp JSON/yelp_academic_dataset_user.json'
csv_path = '/Users/jaeeponde/dsm_project/Yelp JSON/yelp_users.csv'

# Open JSON and CSV files
with open(user_path, 'r') as json_file, open(csv_path, 'w', newline='', encoding='utf-8') as csv_file:
    # Choose the fields you want to extract
    fieldnames = [
        'user_id', 'name', 'review_count', 'yelping_since', 'useful', 'funny',
        'cool', 'elite', 'friends', 'fans', 'average_stars', 'compliment_hot',
        'compliment_more', 'compliment_profile', 'compliment_cute',
        'compliment_list', 'compliment_note', 'compliment_plain',
        'compliment_cool', 'compliment_funny', 'compliment_writer', 'compliment_photos'
    ]

    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()

    for line in json_file:
        user = json.loads(line)
        # Flatten friends (list of user_ids) into count or string if needed
        user['friends'] = len(user.get('friends', '').split(', ')) if user.get('friends') else 0
        writer.writerow({field: user.get(field, None) for field in fieldnames})
```

```python
import json
import csv
import os

# Input path
review_path = '/Users/jaeeponde/dsm_project/Yelp JSON/yelp_academic_dataset_review.json'
output_dir = '/Users/jaeeponde/dsm_project/Yelp JSON/review_chunks'
os.makedirs(output_dir, exist_ok=True)

# Settings
chunk_size = 100000
fieldnames = ['review_id', 'user_id', 'business_id', 'stars', 'useful', 'funny', 'cool', 'text', 'date']

# Process
with open(review_path, 'r') as infile:
    chunk_num = 0
    line_count = 0
    outfile = open(os.path.join(output_dir, f'yelp_reviews_chunk_{chunk_num}.csv'), 'w', newline='', encoding='utf-8')
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for line in infile:
        review = json.loads(line)
        writer.writerow({field: review.get(field, None) for field in fieldnames})
        line_count += 1

        if line_count % chunk_size == 0:
            outfile.close()
            chunk_num += 1
            outfile = open(os.path.join(output_dir, f'yelp_reviews_chunk_{chunk_num}.csv'), 'w', newline='', encoding='utf-8')
            writer = csv.DictWriter(outfile, fieldnames=fieldnames)
            writer.writeheader()

    outfile.close()
```
