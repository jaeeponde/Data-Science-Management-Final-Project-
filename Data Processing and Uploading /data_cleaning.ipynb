```python
import pandas as pd
import glob
import os

# Path to your reviews folder
reviews_folder = '/Users/jaeeponde/dsm_project/Yelp JSON/review_chunks'

# Load restaurant businesses
restaurants_df = pd.read_csv('/Users/jaeeponde/dsm_project/cleaned/restaurants_only.csv')
restaurant_ids = set(restaurants_df['business_id'])  # for fast lookup

# Collect filtered review chunks
filtered_chunks = []

# Get all CSV files in the folder
csv_files = glob.glob(os.path.join(reviews_folder, '*.csv'))

# Process each file
for file in csv_files:
    print(f"Processing {file}...")
    reviews_df = pd.read_csv(file)
    
    # Filter reviews where business_id is in restaurants
    filtered = reviews_df[reviews_df['business_id'].isin(restaurant_ids)]
    
    filtered_chunks.append(filtered)

# Concatenate all filtered reviews
filtered_reviews = pd.concat(filtered_chunks, ignore_index=True)

# Save to output
filtered_reviews.to_csv('restaurant_reviews.csv', index=False)

```

```python
import pandas as pd

# Load the filtered reviews (only required columns to save memory)
reviews_df = pd.read_csv("/Users/jaeeponde/dsm_project/final_csv/sampled_reviews.csv", usecols=["user_id"])
review_user_ids = set(reviews_df["user_id"])
```

```python
import json

core_data = []
friends_data = []

with open("/Users/jaeeponde/dsm_project/Yelp JSON/yelp_academic_dataset_user.json", "r") as f:
    for line in f:
        user = json.loads(line)
        if user["user_id"] in review_user_ids:
            # Store user core info
            core_data.append({
                "user_id": user["user_id"],
                "name": user["name"],
                "review_count": user["review_count"],
                
                "average_stars": user["average_stars"]
            })

            # Store friends (as list of rows: user_id, friend_id)
            for friend_id in user.get("friends", "").split(", "):
                if friend_id:  # skip empty strings
                    friends_data.append({"user_id": user["user_id"], "friend_id": friend_id})

# Convert to DataFrames and save
pd.DataFrame(core_data).to_csv("user_core.csv", index=False)
pd.DataFrame(friends_data).to_csv("user_friends.csv", index=False)
```

```python
import pandas as pd
import os
import numpy as np

# Path to your CSV
input_path = "/Users/jaeeponde/dsm_project/final_csv/sampled_reviews.csv"

# Load the CSV
df = pd.read_csv(input_path)

# Clean the 'review' column
df['text'] = df['text'].str.replace(r'[\\/]', '', regex=True)

# Split into 10 parts
split_dfs = np.array_split(df, 10)

# Save each part
base_dir = os.path.dirname(input_path)
for i, part in enumerate(split_dfs, 1):
    output_path = os.path.join(base_dir, f"sampled_reviews_part_{i}.csv")
    part.to_csv(output_path, index=False)

print("Split and saved successfully.")
```

```python
import pandas as pd
import json

# Path to the CSV file
file_path = '/Users/jaeeponde/dsm_project/final_csv/sampled_reviews.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Select the first 50,000 rows
df_sample = df.head(50000)

# Remove '/' and '\' from all string entries
df_cleaned = df_sample.applymap(lambda x: x.replace('/', '').replace('\\', '') if isinstance(x, str) else x)

# Convert to JSON (list of dicts)
json_data = df_cleaned.to_dict(orient='records')

# Save to a JSON file
with open('/Users/jaeeponde/dsm_project/final_csv/cleaned_reviews.json', 'w') as f:
    json.dump(json_data, f, indent=2)

print("Cleaned data saved to cleaned_reviews.json")
```

```python
import pandas as pd
import json

# Load the list of user_ids from the philly_reviews CSV file
reviews_df = pd.read_csv("/Users/jaeeponde/dsm_project/philadelphia/philly_reviews.csv")
user_ids = reviews_df['user_id'].unique()

# Load the Yelp JSON file
with open("/Users/jaeeponde/dsm_project/Yelp JSON/yelp_academic_dataset_user.json", "r") as f:
    users_data = []
    for line in f:
        user = json.loads(line)
        if user['user_id'] in user_ids:
            users_data.append({
                'user_id': user['user_id'],
                'name': user.get('name', ''),
                'review_count': user.get('review_count', 0),
                'average_stars': user.get('average_stars', 0.0)
            })

# Create a DataFrame from the filtered user data
user_df = pd.DataFrame(users_data)

# Save the DataFrame to a new CSV
user_df.to_csv("/Users/jaeeponde/dsm_project/philadelphia/filtered_users.csv", index=False)

print("Filtered users have been saved to 'filtered_users.csv'.")


```
